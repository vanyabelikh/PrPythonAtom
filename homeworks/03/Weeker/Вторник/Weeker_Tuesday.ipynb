```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
columns = ['Age', 'Workclass', 'Fnlwgt', 'Education', 'Education_num', 'Marital_status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Capital_gain', 'Capital_loss', 'Hours_per_week', 'Native_country', 'Param']
dfTrades = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', ', ', header = None)
dfTrades.columns = columns
dfTrades
```
