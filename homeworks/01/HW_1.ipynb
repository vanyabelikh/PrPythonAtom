```python
class HashMap:
    class Entry:
        def __init__(self, key, value):
            self.key = key
            self.value = value

        def get_key(self):
            return self.key

        def get_value(self):
            return self.value

        def __eq__(self, other):
            if self.key == other.key:
                return True
            return False

    def __init__(self, bucket_num=64):

        self.capasity = bucket_num
        self.size = 0
        self.map = [HashMap.Entry(None, None)] * bucket_num

    def get(self, key, default_value=None):

        index = self._get_index(self._get_hash(key))
        flag = index
        flag2 = 0
        if self.map[index] != HashMap.Entry(None, None):
            while hash(key) != hash(self.map[index].key):
                index += 1
                if index >= self.capasity:
                    index = 0
                if index == flag:
                    flag2 = 1
                    break
            if not flag2: return self.map[index].get_value()
        return default_value

    def put(self, key, value):

        if self.size > (2 / 3) * self.capasity:
            self.capasity *= 2
            new_map = [HashMap.Entry(None, None)] * self.capasity
            for i in self.map:
                if i != HashMap.Entry(None, None):
                    index = hash(i.get_key()) % self.capasity
                    while new_map[index] != HashMap.Entry(None, None):
                        if new_map[index] == i:
                            return
                        index = (index + 1) % self.capasity
                    new_map[index] = i
            self.map = new_map

        # print(hash(key)%self.capasity)
        item = HashMap.Entry(key, value)
        index = self._get_index(self._get_hash(key))

        if self.map[index] == HashMap.Entry(None, None):
            self.map[index] = item
            self.size += 1
            return
        while self.map[index] != HashMap.Entry(None, None):
            if self.map[index].key == item.key:
                self.map[index] = item
                return
            index = (index + 1) % self.capasity
        self.map[index] = item
        self.size += 1
        return

    def __len__(self):

        return self.size

    def _get_hash(self, key):

        return hash(key)

    def _get_index(self, hash_value):

        return hash_value % self.capasity

    def __str__(self):

        out = ''
        for i in range(0, self.capasity):
            out += ('   ({}, {})   '.format(self.map[i].key, self.map[i].value))
        return out

    def values(self):

        for i in self.map:
            if i != HashMap.Entry(None, None):
                yield i.value

    def keys(self):

        for i in self.map:
            if i != HashMap.Entry(None, None):
                yield i.key

    def items(self):

        for i in self.map:
            if i != HashMap.Entry(None, None):
                yield (i.key, i.value)
```

```python
class HashSet(HashMap):

    def get(self, key, default_value=None):
        super().get(key, default_value)

    def put(self, key, value):
        super().put(key, None)

    def __len__(self):
        return super().__len__()

    def values(self):
        return super().keys()
```
